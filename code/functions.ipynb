```python
# Definition
def times(x, y): # Create and assign function
    return x * y # Body executed when called

times = lambda x, y: x * y

# Calls
times(2, 4)
x = times(3.14, 4) # Save the result object
times('Py', 4) # Functions are "typeless"

# Polymorphism in Python
times('not', 'quite')
```
